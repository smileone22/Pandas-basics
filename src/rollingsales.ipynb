# 8. Bin the prices of properties sold

```python
# use the results from the summary statistics calculated in the previous steps to come up with minimally 5 bins to fit the prices in
# ⚠️ bin the prices and create a visualization that compares the bins

# use chapter 7 in Data Analysis in Python to do this… or use the slides on binning and/or the docs on pd.cut
# hint: the sort_index method, and the keyword argument, asecending = True or False may be helpful


data['bins']=pd.cut(data['SALE PRICE'],[0,300000,1000000,3500000,10000000,2155000000],labels=['a','b','c','d','e'],include_lowest=True)
d2= data[['SALE PRICE','bins']]
d3= d2.groupby(by='bins').agg(['count','max','min','mean'])
```

```python
d3
```

```python
vals= d2.groupby(by='bins').mean()['SALE PRICE'].tolist()
```

```python
vals
```

```python
import matplotlib.pyplot as plt
# labels
bins = ['a','b','c','d','e']
# y values
meanvals= vals

colors = {'a: (0, 500000]':'red', 'b: (500000, 2000000]':'blue', 'c: (2000000, 10000000]':'purple', 'd: (10000000, 100000000]':'green', 'e: (100000000, 2155000000]':'lavender'}         
labels = list(colors.keys())
handles = [plt.Rectangle((0,0),1,1, color=colors[label]) for label in labels]
plt.legend(handles, labels,title="BIN price range")

# the x values (based on number of labels)
x = np.arange(len(bins))
plt.bar(x, meanvals, align='center', color=['red', 'blue', 'purple', 'green', 'lavender'])
plt.rcParams['figure.facecolor'] = 'white'
plt.xticks(x,bins, fontsize=10)
plt.ylim(0,  50000000)
plt.ylabel('bins')
plt.xlabel('Average Sale Price')
plt.title('Average Sale Price in 5 price bins')
plt.figure(facecolor="yellow")
plt.show()
```
