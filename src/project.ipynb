# 3. Using the Data

* In previous assignment, analysis of this data was used in 
    1) finding sidewalk cafés with the maximum space 
    2) finding the city in New York has the most active sidewalk cafés

```python
round(df.describe()) #calculate summary statistics(1)
```

```python
# create a new calculated column
# Guessing the number of chairs allocated per table calculated through SWC_CHAIRS/SWC_TABLES
# create a new calculated column(2)
df['chair_per_table']=round(df['SWC_CHAIRS']/df['SWC_TABLES'])
```

```python
# create a new calculated column
# If true, the amount of square footage space for its sidewalk café the business applied for matches the  size of the sidewalk café space by square footage as permitted by the business’s DCA license.
# create a new calculated column(3)
df['application_sqft_success']=df['SWC_SQ_FT']==df['APP_SQ_FT']
```

```python
df[df['application_sqft_success']==False] # sidewalk café that did not get sqft size as they applied for 
```

```python
df[df['LICENSE_NBR'].isna()==True]
```

```python
# finding sidewalk cafés with the maximum space
df[df['SWC_SQ_FT']==df['SWC_SQ_FT'].max()]
```

```python
# calculate value counts (4)
df.groupby(by='CITY').size().sort_values()
```

```python
actives= df[df['LIC_STATUS']=='Active']
```

```python
# finding the city in New York has the most active sidewalk cafés
actives.groupby(by='CITY').size().sort_values(ascending=False)
```

```python
# finding sidewalk café with most chair per table
df[df['chair_per_table']==df['chair_per_table'].max()]
```

```python
# top five streets with most sidewalk cafés
# calculate value counts (5)
df.groupby(by='STREET').size().sort_values(ascending=False).head(5)
```

```python
# calculate value counts (6)
df.groupby(by='LIC_STATUS').size().sort_values(ascending=False)
```

```python
# calculate value counts + summary_stat(7)
df.groupby(by='LIC_STATUS').agg(['mean', 'count','std'])
```

```python
# calculate value counts + summary_stat(8)
df.groupby(by='SWC_TYPE').agg(['mean', 'count'])
```
